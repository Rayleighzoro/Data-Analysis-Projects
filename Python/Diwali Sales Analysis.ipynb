# **Diwali Sales Analysis**

```python
# import all important libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import the data into a new variable called df
df = pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')

# Show 
df.shape
```

```python
df.dtypes
```

```python
# Check the top 5 lines of the file to see the data is loaded successfully
df.head()
```

```python
# Check the info fo the available dataframe.
df.info()
```

```python
# Drop the unnecessary rows 
df.drop(['Status', 'unnamed1'], axis = 1, inplace = True)
```

```python
df.info()
```

```python
# Check if the data frame has any null value 
pd.isnull(df['Amount']).sum()
```

```python
# Remove the null values from the dataframe using the dropna function 
df.dropna(inplace = True)
```

```python
# Check if it has removed the null values
pd.isnull(df).sum()
```

```python
# Check and change the data type of for further analysis
df.info()
```

```python
# Change the Amount data type to integer.
df['Amount'] = df['Amount'].astype('int')
```

```python
df.info()
```

```python
# Rename columns as required
df.rename(columns = {'Marital_Status': 'Shaadi'})
```

# **Exploratory Data Analysis (EDA)**

# **Gender** 

```python
# Set figure size
plt.figure(figsize=(6, 4))

# Create countplot
ax = sns.countplot(x='Gender', data=df, hue = 'Gender')

# Add bar labels
for bars in ax.containers:
    ax.bar_label(bars)

# Add title and axis labels
ax.set_title('Distribution of Gender', fontsize=16, fontweight='bold')
ax.set_xlabel('Gender', fontsize=14)
ax.set_ylabel('Count', fontsize=14)

# Customize tick label size
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Improve layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
Sales_gen = df.groupby(['Gender'])['Amount'].sum().reset_index().sort_values(by = 'Amount', ascending= False)

# Set figure size
plt.figure(figsize = (5,3))

# Create barplot
ax = sns.barplot(x = 'Gender', y = 'Amount', data = Sales_gen, hue = 'Gender')

    
# Add title and axis labels
plt.title('Distribution of Gender as Amount spend', fontsize=16, fontweight='bold')
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Amount', fontsize=14)
```

### Age Group 


```python
# Adjust the size as per your desire
plt.figure(figsize = (10, 5 ))

# data plot and assigning
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

# Add title for the chart
plt.title('Gender-Wise Distribution Across Age Groups')

# Add bar labels
for bars in ax.containers:
    ax.bar_label(bars)
```

# **States**

**Total number of Orders from top 10 states**

```python
# Group and sort the data
sales_state = df.groupby(['State'])['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

# Set figure size
plt.figure(figsize=(10, 5))

# Define a list of colors (same length as number of bars)
custom_colors = sns.color_palette("husl", len(sales_state))  # Or use your own list of hex colors

# Create barplot with different colors
sns.barplot(data=sales_state, x='State', y='Orders', hue = 'State')

# improve label readability
plt.xticks(rotation=15)
plt.title('Total Orders by State')
plt.tight_layout()

plt.show()
```

**Total amount/ sales from top 10 States**

```python
## Group by State the Amount gathered from each state
State_Sal = df.groupby(['State'])['Amount'].sum().reset_index().sort_values(by='Amount', ascending= False).head(10)

# Set a customized size for the chart 
plt.figure(figsize = (12, 5))

# Visualization
sns.barplot(data = State_Sal, x = 'State', y = 'Amount', hue = 'State')

#improve label readability
plt.xticks(rotation = 20)
plt.title('Total Amount Sales by State')
plt.show()
```

# **Marital Status**

```python
# Count the Marital Status and create a bar chart
ax = sns.countplot(data = df, x = 'Marital_Status')

# Asign the title to the chart
plt.title('Distribution of Marital Status count')

# Create a size and figure 
plt.figure(figsize = (10,5))

# Create a container to store the count and show in the bar
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Group by Marital Status and Gender as per the Amount
Sales_Marital_Status = df.groupby(['Marital_Status','Gender'])['Amount'].sum().reset_index().sort_values(by='Amount', ascending= False).head(10)

# Create a visualization for this new data set 
sns.barplot(data = Sales_Marital_Status, x = "Marital_Status", y ="Amount" , hue = 'Gender')

# Assign a title for the chart 
plt.title("Marital Status and Gender Amount")

# Assign the size for the chart 
plt.figure(figsize=(10,5))
```

# **Occupation**

```python
# Assign a customized size for the chart
plt.figure(figsize = (20,5))

# Assign a title for the Chart 
plt.title("Count of orders as per the Occupation")

# assign the data to be shown in the chart
ax = sns.countplot(data = df, x = 'Occupation', hue = 'Occupation')

# Create a container for storing the count
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Group by occupation and sort the value as per the amount
Sales_Occupation = df.groupby(['Occupation']) ['Amount'].sum().reset_index().sort_values(by='Amount', ascending= False).head(10)

# Assign the size for the chart 
plt.figure(figsize = (15, 5))

# Assign the title for the chart 
plt.title("Sales as per Occupation")

# Create a chart using the seaborn 
sns.barplot(data = Sales_Occupation , x = 'Occupation', y = 'Amount', hue = 'Occupation')

plt.show
```

# **Product Category**

```python
plt.figure(figsize = (20, 5))

ax = sns.countplot(data = df, x = 'Product_Category', hue = 'Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
df.groupby(['Product_Category'])['Amount'].sum().sort_values( ascending = False).reset_index().head(10)
```

```python
Pro_Sales = df.groupby(['Product_Category'])['Amount'].sum().sort_values( ascending = False).reset_index().head(10)

plt.figure(figsize = (20,5))
plt.title("Sales by Product Category")

sns.barplot(data = Pro_Sales, x = 'Product_Category', y = 'Amount', hue = 'Product_Category')
plt.show()
```

```python
Pro_Sales = df.groupby(['Product_ID'])['Amount'].sum().sort_values( ascending = False).reset_index().head(10)

plt.figure(figsize = (20,5))
plt.title("Sales by Product ID")

sns.barplot(data = Pro_Sales, x = 'Product_ID', y = 'Amount', hue = 'Product_ID')
plt.show()
```
